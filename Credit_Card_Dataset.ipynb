```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
creditcard1 = pd.read_csv("/content/CC GENERAL.csv")

creditcard1['MINIMUM_PAYMENTS'].fillna(median_value, inplace=True)
creditcard1['CREDIT_LIMIT'].fillna(median_value, inplace=True)

# Select numeric columns
numeric_columns = creditcard1.select_dtypes(include=['float64', 'int64']).columns
print("Numeric columns:", numeric_columns)

# Standardize the numeric columns
scaler = StandardScaler()
creditcard1[numeric_columns] = scaler.fit_transform(creditcard1[numeric_columns])

# Perform the elbow method to find the optimal number of clusters
inertia = []  # List to store the inertia values for each k

# Loop through different values of k (1 to 10 clusters in this case)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=77)
    kmeans.fit(creditcard1[numeric_columns])  # Fit KMeans on the standardized data
    inertia.append(kmeans.inertia_)  # Append the inertia value for each k

# Plot the elbow curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.show()
```

#3: Conducting Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA

# Initialize PCA with 10 components
pca = PCA(n_components=10)

# Perform PCA on the standardized numeric columns
pca_result = pca.fit_transform(creditcard1[numeric_columns])

# Create a DataFrame to hold the PCA components
pca_df = pd.DataFrame(data=pca_result, columns=[f'PCA{i+1}' for i in range(10)])

# Add the 'CUST_ID' and 'Cluster' columns to the PCA DataFrame
pca_df['CUST_ID'] = creditcard1['CUST_ID']
pca_df['Cluster'] = creditcard1['Cluster']

# Print the first few rows of the PCA DataFrame
print(pca_df.head())
```

```python
display(pca_df)
```

```python
# Retrieve the loadings (weights) for PCA1 to PCA4
pca_loadings = pd.DataFrame(pca.components_[:4].T,  # Transpose to align features with PCs
                            columns=['PC1', 'PC2', 'PC3', 'PC4'],  # Columns are the principal components
                            index=numeric_columns)  # Use the numeric feature names as the index

# Display the loadings
print("PCA Loadings (Feature Contributions to Each Principal Component):")
import numpy as np
from IPython.display import display
display(pca_loadings)
```

```python
# Retrieve the loadings (weights) for PCA1 only
pca_loadings_pc1 = pd.DataFrame(pca.components_[0].T,  # Get the first principal component's loadings
                                columns=['PC1'],  # The column name for PC1
                                index=numeric_columns)  # Use the numeric feature names as the index

# Display the loadings for PC1
print("PCA1 Loadings (Feature Contributions to PC1):")
import numpy as np
from IPython.display import display
display(pca_loadings_pc1)
```

```python
# Retrieve the loadings (weights) for PCA1 and PCA2
pca_loadings_pc1 = pd.DataFrame(pca.components_[0].T,  # Get the first principal component's loadings
                                columns=['PC1'],  # The column name for PC1
                                index=numeric_columns)  # Use the numeric feature names as the index

pca_loadings_pc2 = pd.DataFrame(pca.components_[1].T,  # Get the second principal component's loadings
                                columns=['PC2'],  # The column name for PC2
                                index=numeric_columns)  # Use the numeric feature names as the index

# Combine both PCA1 and PCA2 loadings
pca_loadings = pd.concat([pca_loadings_pc1, pca_loadings_pc2], axis=1)

# Display the loadings for PCA1 and PCA2
print("PCA1 and PCA2 Loadings (Feature Contributions):")
import numpy as np
from IPython.display import display
display(pca_loadings)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the dataset
creditcard1 = pd.read_csv("/content/CC GENERAL.csv")

# Impute missing values for 'MINIMUM_PAYMENTS' and 'CREDIT_LIMIT' with the median
median_value = creditcard1['MINIMUM_PAYMENTS'].median()
creditcard1['MINIMUM_PAYMENTS'].fillna(median_value, inplace=True)

median_value = creditcard1['CREDIT_LIMIT'].median()
creditcard1['CREDIT_LIMIT'].fillna(median_value, inplace=True)

# Select numeric columns
numeric_columns = creditcard1.select_dtypes(include=['float64', 'int64']).columns
print("Numeric columns:", numeric_columns)

# Standardize the numeric columns
scaler = StandardScaler()
creditcard1[numeric_columns] = scaler.fit_transform(creditcard1[numeric_columns])

# Perform KMeans clustering with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=77)
creditcard1['Cluster'] = kmeans.fit_predict(creditcard1[numeric_columns])

# Adjust cluster labels to start from 1 instead of 0
creditcard1['Cluster'] = creditcard1['Cluster'] + 1

# Perform PCA with 4 components
pca_4 = PCA(n_components=4)
pca_result_4 = pca_4.fit_transform(creditcard1[numeric_columns])

# Create a new DataFrame to store the 4 principal components
pca_columns_4 = [f'PC{i+1}' for i in range(4)]  # Naming the components as PC1, PC2, PC3, PC4
creditcard4 = pd.DataFrame(pca_result_4, columns=pca_columns_4)

# Add the non-numeric columns back to the PCA result (if needed)
non_numeric_columns = creditcard1.select_dtypes(exclude=['float64', 'int64']).columns
creditcard4 = pd.concat([creditcard1[non_numeric_columns].reset_index(drop=True), creditcard4], axis=1)

# Combine PCA and Clustering results (Add PCA components and cluster labels to original data)
creditcard4['Cluster'] = creditcard1['Cluster']

# Step 1: Focus on PCA1 and create bar charts for each cluster
plt.figure(figsize=(10,6))

# Group by Cluster and calculate the mean of PCA1
pca1_cluster_means = creditcard4.groupby('Cluster')['PC1'].mean()

# Plot the bar chart for PCA1 by cluster
pca1_cluster_means.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen', 'orange'])
plt.title('Mean PCA1 by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Mean PCA1')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load the original dataset
creditcard = pd.read_csv("/content/CC GENERAL.csv")

# Impute missing values for 'MINIMUM_PAYMENTS' and 'CREDIT_LIMIT' with the median
creditcard['MINIMUM_PAYMENTS'].fillna(creditcard['MINIMUM_PAYMENTS'].median(), inplace=True)
creditcard['CREDIT_LIMIT'].fillna(creditcard['CREDIT_LIMIT'].median(), inplace=True)

# Select numeric columns (excluding the identifier column)
numeric_columns = creditcard.select_dtypes(include=['float64', 'int64']).columns

# Standardize the numeric columns
scaler = StandardScaler()
standardized_data = scaler.fit_transform(creditcard[numeric_columns])

# Perform KMeans clustering with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=77)
clusters = kmeans.fit_predict(standardized_data)

# Add the Cluster column to the original dataset
creditcard['Cluster'] = clusters + 1  # Adjust cluster labels to start from 1

# Display the first few rows of the updated dataset
display(creditcard.head())

# Optionally save the updated dataset
creditcard.to_csv("/content/CC_with_Clusters.csv", index=False)
```
